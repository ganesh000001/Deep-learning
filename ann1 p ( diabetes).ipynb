```python
import numpy as np
import pandas as pd
import tensorflow
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data = pd.read_csv('F:/Social Prachar/Raj sir/july/14 july 2021/diabetes.csv',header=None,skiprows=1)
```

```python
data.head()
```

```python
X = data.iloc[:,0:8].values
y = data.iloc[:,8].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)
```

```python
model = Sequential()
model.add(Dense(32,input_dim=8,activation='sigmoid'))
model.add(Dense(18,activation='tanh'))

model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history= model.fit(X_train,y_train,epochs=100,verbose=2)
```

```python
y_pred = model.predict(X_test)
#Converting predictions to label
pred = list()
for i in range(len(y_pred)):
    pred.append(np.argmax(y_pred[i]))
#Converting one hot encoded test label to label
test = list()
for i in range(len(y_test)):
    test.append(np.argmax(y_test[i]))
```

```python
from sklearn.metrics import accuracy_score
```

```python
a = accuracy_score(pred,test)
print('Accuracy is :',a*100)
```

```python
history = model.fit(X_train,y_train, validation_data = (X_test,y_test),epochs=100,verbose=2)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss']) 
plt.plot(history.history['val_loss']) 
plt.title('Model loss') 
plt.ylabel('Loss') 
plt.xlabel('Epoch') 
plt.legend(['Train', 'Test'], loc='upper left') 
plt.show()
```
